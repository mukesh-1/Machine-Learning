```python
import numpy as np
import pandas as pd
import seaborn as sns
import sklearn.model_selection as skm
import sklearn.linear_model as skl
```

```python
dataset = pd.read_csv('loanpredictiontrainingset.csv')
dataset.shape
```

```python
dataset.isna().sum()
```

```python
dataset=dataset.dropna()
```

```python
dataset
```

```python
dataset['Dependents'] = dataset['Dependents'].replace(to_replace = '3+',value=4)
```

```python
dataset['Dependents'].value_counts()
```

```python
dataset.head()
```

```python
dataset.drop('level_0',axis=1,inplace=True)
```

```python
dataset.head()
```

```python
dataset.replace({'Gender':{'Male' : 1, 'Female' : 0}, 'Married' : {'Yes' : 1,'No':0}, 'Education' : {'Graduate' : 1,'Not Graduate':0},
                'Self_Employed':{'Yes':1,'No':0},'Property_Area':{'Rural':0,'Urban':1,'Semiurban':2}},inplace=True)
```

```python
dataset.head()
```

```python
dataset['Dependents'] = dataset['Dependents'].astype('int')
```

```python
dataset.replace({'Loan_Status' : {'Y' : 1,'N':0}},inplace=True)
```

```python
x=dataset.iloc[:,2:-1].values
x[0]
```

```python
dataset.head()
```

```python
y=dataset.iloc[:,-1].values
```

```python
x_train,x_test,y_train,y_test=skm.train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
log_regression = skl.LogisticRegression()
log_regression.fit(x_train,y_train)
```

```python
log_y_pred = log_regression.predict(x_test)
```
